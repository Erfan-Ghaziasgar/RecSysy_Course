حالا نوبت به آن رسیده است که داده‌های <code>object</code> در درون لیست را مشاهده کنیم و به شکل نرمالی دست پیدا کنیم.<br>
      برای اینکار چند نمونه‌ از داده‌های موجود در لیست‌ها را مشاهده کنید و با توجه به دیدگاه خود تابع <code>preprocessing_text</code> رو پیاده سازی کنید و داده‌های دیتافریم را نرمال کنید.

```python
object_columns = df.select_dtypes(include=['object']).columns
random_state = 42
# display 5 random rows
df[object_columns].sample(5, random_state=random_state)
```

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re

lemma = WordNetLemmatizer()

stop_words = set(stopwords.words("english"))
def text_preprocess(text):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", "", text)
    words = text.split()
    words = [word for word in words if word not in stop_words]
    words = [lemma.lemmatize(word) for word in words]
    words = [word for word in words if word != ""]
    text = " ".join(words)
    return text
```

<h1 dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
  <font face="vazir" color="#0099cc" size=5>
    گام چهارم: بردارسازی نمایه‌ها
  </font>
</h1>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
  <font face="vazir" size=3>
برای استفاده از داده‌های متنی ابتدا از طریق تابع پیش‌پردازشی که برای متن‌ها ساختیم ستون‌های متنی مورد نیاز را پیش‌پردازش می‌کنیم و سپس حاصل این پردازش‌ها را با یکدیگر ترکیب کرده تا ویژگی‌ جدیدی ایجاد کنیم (<code>text_feature</code>) تا بتوانیم از با پردازش کمتر داده‌های متنی رو به داده‌های عددی و برداری تبدیل کنیم.<br>
      پس از انجام عملیات پیش‌پردازش و ترکیب ویژگی‌های متنی (منظور جمع رشته‌ای متن‌هاست) با استفاده از روش <code>TF-IDF</code> بردارسازی را روی ستون‌های متنی اعمال می‌کنیم تا بعد با ترکیب این بردار‌ها با مقادیر عددی به راحتی شباهت کسینوسی آن‌ها را دریابیم..</p>

```python
# TO DO
from sklearn.feature_extraction.text import TfidfVectorizer

df['artists'] = df['artists'].apply(lambda x: text_preprocess(x) if isinstance(x, str) else '')
df['track_genre'] = df['track_genre'].apply(lambda x: text_preprocess(x) if isinstance(x, str) else '')

text_features = df['track_name'] + ' ' + df['artists'] + ' ' + df['track_genre']
text_features = text_features.fillna('')  # Fill NaN values with empty strings
text_features = text_features.astype(str)  # Ensure all values are strings
text_features = text_features.apply(lambda x: x.strip())  # Remove leading/trailing spaces
vectorizer = TfidfVectorizer()
X_text = vectorizer.fit_transform(text_features)
```

```python
# TO DO
from sklearn.preprocessing import MinMaxScaler

numeric_cols = df.select_dtypes(include=[np.number]).columns
scaler = MinMaxScaler()
X_numeric = scaler.fit_transform(df[numeric_cols])
```

```python
from scipy.sparse import hstack

X_combined = hstack([X_text, X_numeric])
```

<h1 dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
  <font face="vazir" color="#0099cc" size=5>
    گام هفتم: شباهت سنجی
  </font>
</h1>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
  <font face="vazir" size=3>
در این مرحله باید با استفاده <stronge>شباهت کسینوسی</stronge> محاسبه شباهت آیتم‌ها بپردازیم.
  </font>
</p>

```python
from sklearn.metrics.pairwise import cosine_similarity
# TO Do
df['feature_vector'] = list(X_combined.toarray())
similarity_matrix = cosine_similarity(X_combined)
```

<h1 dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
  <font face="vazir" color="#0099cc" size=5>
    گام هشتم: پیاده‌سازی تابع پیشنهاددهنده
  </font>
</h1>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
  <font face="vazir" size=3>
تابع <code>recommender</code> به گونه‌ای پیاده‌سایز کنید که ۲۱ آیتم اولی که مشابه هستند را در قالب یک لیست برگرداند.
  </font>
</p>

```python
def recommender(music_name):
    if music_name not in df['track_name'].values:
        return "Music not found in the dataset."
    
    music_index = df[df['track_name'] == music_name].index[0]
    similarity_scores = list(enumerate(similarity_matrix[music_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    
    recommended_indices = [i[0] for i in similarity_scores[1:22]]  # Get top 21 recommended tracks (excluding the first one which is the same track)
    recommended_tracks = df.loc[recommended_indices, 'track_name'].tolist()
    
    return recommended_tracks
```

  <p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">

```python
selection = ['Ilahi', 'Loony', 'This Is Your Life', 'Solamente una Vez', 'In Da Getto']
```

```python
recommendation_dict = {}
for track in selection:
    recommendation_dict[track] = recommender(track)

    if recommendation_dict[track] == "Music not found in the dataset.":
        print(f"Track '{track}' not found in the dataset.")
    else:
        print(f"Recommendations for '{track}': {recommendation_dict[track]}")

submission = pd.DataFrame.from_dict(recommendation_dict, orient='index').transpose()
submission
```
