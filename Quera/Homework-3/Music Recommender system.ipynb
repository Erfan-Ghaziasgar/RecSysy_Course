```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re

lemma = WordNetLemmatizer()

stop_words = set(stopwords.words("english"))
def text_preprocess(text):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", "", text)
    words = text.split()
    words = [word for word in words if word not in stop_words]
    words = [lemma.lemmatize(word) for word in words]
    words = [word for word in words if word != ""]
    text = " ".join(words)
    return text
```

```python
def recommender(music_name):
    if music_name not in df['track_name'].values:
        return "Music not found in the dataset."
    
    music_index = df[df['track_name'] == music_name].index[0]
    similarity_scores = list(enumerate(similarity_matrix[music_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    
    recommended_indices = [i[0] for i in similarity_scores[1:22]]  # Get top 21 recommended tracks (excluding the first one which is the same track)
    recommended_tracks = df.loc[recommended_indices, 'track_name'].tolist()
    
    return recommended_tracks
```
